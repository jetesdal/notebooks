```python
data_path = 'mpiom/'

filenames = {'U': data_path + 'foo3.fixed.nc',
             'V': data_path + 'foo3.fixed.nc'}

variables = {'U': 'uko',
             'V': 'vke'}

dimensions = {'lon': 'lon', 'lat': 'lat', 'time': 'time'}
```

```python
fieldset = FieldSet.from_netcdf(filenames, variables, dimensions, allow_time_extrapolation=True,full_load=True)
```

```python
np.shape(fieldset.U.data)
```

```python
from mpl_toolkits.basemap import Basemap
```

```python
plt.figure(figsize=(12,7))

m = Basemap(projection='cyl', llcrnrlat=48, urcrnrlat=70, llcrnrlon=-65, urcrnrlon=-45, 
            lat_ts=0, resolution='l')
m.fillcontinents(color='lightgrey')
m.drawcoastlines()
m.drawparallels(np.arange(-60,90,2),labels=[1,0,0,0],fontsize=10)
m.drawmeridians(np.arange(0,360,2),labels=[0,0,0,1],fontsize=10)

plt.show()
```

```python
# Construct a grid of particles at the source region (here: Labrador Sea)
lonp, latp = np.meshgrid(range(-60, -52), range(56, 64))
```

```python
# Alternatively, we can start 'n' particles on a meridional line
npart = 10
lonp = -56 * np.ones(npart)
latp = [i for i in np.linspace(56, 64, npart)]
```

```python
def DeleteParticle(particle, fieldset, time, dt):
    particle.delete()
```

```python
pset = ParticleSet.from_list(fieldset, JITParticle, lon=lonp, lat=latp)
```

```python
pset.show()
```

```python
print(int(mpiom.time[-1]))
```

```python
for cnt in range(int(mpiom.time[-1]/15)):
    # First plot the particles
    pset.show(savefile='mpiom_particles_1x1/LabSea'+str(cnt).zfill(3), field='vector', land=True, vmin=0.0, vmax=0.5)

    # Then advect the particles for 15 days
    pset.execute(AdvectionRK4, runtime=delta(days=15), dt=delta(days=1),
                 recovery={ErrorCode.ErrorOutOfBounds: DeleteParticle})
```
